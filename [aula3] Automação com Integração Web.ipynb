```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
#vamos inserir o valor de webdriver do navegador web Chrome na variável nav
nav = webdriver.Chrome()
# pesquisar cotação do dólar
nav.get("https://www.google.com/")
# vamos colocar o código HTML que indica o elemento da caixa de pesquisa para o programa buscar no navegador (caminho xpath)
nav.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("Cotação do Dólar")
nav.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)                                                                           
cotacao_dolar = nav.find_element_by_xpath('/html/body/div[7]/div/div[9]/div[1]/div/div[2]/div[2]/div/div/div[1]/div/div/div/div/div[1]/div/div[1]/div[1]/div[2]/span[1]').get_attribute('data-value')
print(cotacao_dolar)

# cotação euro
nav = webdriver.Chrome()
nav.get("https://www.google.com/")
nav.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("Cotação do Euro")
nav.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)                                                                           
cotacao_euro = nav.find_element_by_xpath('/html/body/div[7]/div/div[9]/div[1]/div/div[2]/div[2]/div/div/div[1]/div/div/div/div/div[1]/div/div[1]/div[1]/div[2]/span[1]').get_attribute('data-value')
print(cotacao_euro)

# cotação do ouro
nav.get("https://www.melhorcambio.com/") # acessar site
aba_original = nav.window_handles[0] #armazena que a aba original é a 0
nav.find_element_by_xpath('/html/body/div[15]/div[2]/div/table[2]/tbody/tr[2]/td[2]/a/img').click() # pega caminho full xpath do ícone da cotação do ouro e pede para clicar nele
# para trocar de aba, vamos informar de onde viemos e para onde vamos
aba_nova = nav.window_handles[1]
nav.switch_to.window(aba_nova)
cotacao_ouro = nav.find_element_by_id('comercial').get_attribute('value')
cotacao_ouro = cotacao_ouro.replace(",", ".")
print(cotacao_ouro)

nav.quit()

#importar pandas para importar planilha

import pandas as pd
produtos_df = pd.read_excel(r"C:\Users\conta\Documents\Educação\2021 - Desenvolvimento de Software\Curso Hashtag Programação - Python, dados, automação\Aula 3 - Automação com integração web - Web-scraping\Produtos.xlsx")
display(produtos_df)
# modificar o dataframe produtos
# método .loc seleciona apenas as linhas com "Dólar na coluna Moeda
# ao encontrar linhas escrito "Dólar", o .loc altera o valor na coluna "Cotação"
produtos_df.loc[produtos_df['Moeda']=="Dólar", "Cotação"]=float(cotacao_dolar)
produtos_df.loc[produtos_df['Moeda']=="Euro", "Cotação"]=float(cotacao_euro)
produtos_df.loc[produtos_df['Moeda']=="Ouro", "Cotação"]=float(cotacao_ouro)
display(produtos_df)

#alterar valor de preço base e preço final

produtos_df['Preço Base Reais'] = produtos_df['Cotação'] * produtos_df["Preço Base Original"]
produtos_df['Preço Final'] = produtos_df['Preço Base Reais'] * produtos_df['Margem']
display(produtos_df)

# exportar nova planilha sem a coluna com o índice de linhas (, index=False)
produtos_df.to_excel(r"C:\Users\conta\Documents\Educação\2021 - Desenvolvimento de Software\Curso Hashtag Programação - Python, dados, automação\Aula 3 - Automação com integração web - Web-scraping\Produtos Atualizado.xlsx", index=False)
```
